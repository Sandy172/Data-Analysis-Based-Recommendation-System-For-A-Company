```python
from sklearn.model_selection import train_test_split
```

```python
X=DA[['Average Time on App (Minutes/Day)','Average Time on Website (Minutes/Day)','Total Purchase from App (Rupees/year)','Total Purchase from Website (Rupees/year)','Length of Membership (Year)']]
y=DA['Ratings given by customer (Out of 5)']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=500)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```
