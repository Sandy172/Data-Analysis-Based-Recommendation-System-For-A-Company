```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
%matplotlib inline
```

```python
DB=pd.read_csv('Dataset B.csv')
```

```python
DB
```

```python
DB.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=DB[['Buy 1 get 1 Free','Buy 2 get 1 Free','30% off for limited period','40% off for limited period','Shop 5000 rupees and get 200 rs off','Shop 8000 rupees and get 500 rs off','Shop 5000 rupees and get coupon of 1000 rs','Shop 8000 rupees and get coupon of 2000 rs','30% off for 4-year member','40% off for 5-year member']]
y=DB['Scratch Card']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=500)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```
